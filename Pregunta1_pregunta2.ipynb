```python

import pandas as pd
df=pd.read_csv(r'C:\Users\LighthouseLLC\Desktop\Econ\_UNI_Python\data\data\TraficoWeb.csv')
df.columns
```

```python
import re
dir(re)
output1=re.search("[a-z][a-z].wikipedia.org", df.Page[500])
output1[0][0:2] #dos primeras componentes
```

```python
def MakeCountry(p):  #p= page
    output1= re.search("[a-z][a-z].wikipedia.org", p)
    if output1:
        return output1[0][0:2]
    return None

MakeCountry(df.Page[13332])


```

```python

df["Country"]=df["Page"].map(MakeCountry)
df.Country.unique()
df.Country.isna().sum()

df.groupby("Country")["Page"].count().sort_values(ascending=False)
```

```python
df2=df.drop('Cantidad de NaNs',1)
```

```python
#Cantidad de valores nan para cada observación (fila)

df2=df.set_index('Page')
df2.drop(columns=['Country', 'Cantidad de NaNs'], axis=1, inplace=True)
df3=df2.T
df4=df3.isna().sum()
df5=df3.T
df5['NanS']=pd.DataFrame(df4)
df5
```

```python
#Number of Pages with no nulls 
df5[df5['NanS']==0].shape
```

```python
#Muestra 1: Tabla
df6=df5[df5['NanS']==0]
del df6['NanS']
df7=df6.sample(20)
df8=df7.T
df8
```
